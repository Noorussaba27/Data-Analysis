# CORRELATION AND CAUSATION

**Correlation**: A major of the extent of interdependence between variables

**Causation**: The relationship between cause and effect between two variables

**Pearson Correlation**: It measures linear dependence between two variables, with values ranging from -1 to 1, inclusive where +1 represents strong positive correlation while -1 represent strong negative correlation.

How to quantify the significance of the correlation estimate? **p-value**

The p-value is the probability value that the relationship between two variables in statistically significant. Normally we choose a significance lebvel of 0.05 which means that we are 95% sure that the correlation is statistically significant. 
By convention, when

 *p-value is  <  0.001: we say there is strong evidence that the correlation is significant
 
 *the p-value is  <  0.05: there is moderate evidence that the correlation is significant
 
 *the p-value is  <  0.1: there is weak evidence that the correlation is significant.
 
 *the p-value is  >  0.1: there is no evidence that the correlation is significant.

```python
#p-value can easily be calculated using scipy library's stats module
from scipy import stats
```

# MODEL EVALUATION AND REFINEMENT

We will be doing a lot of plotting so it is better to create plotting functions

```python
#Distribution Plots
def DistributionPlot(RedVariable, BlueVariable, RedName, BlueName, Title):
    height=10
    widht=12
    plt.figure(figsize=(width, height))
    
    ax1= sns.distplot(RedVariable, hist=False, color='r', label=RedName)
    ax2=sns.distplot(BlueVariable, hist=False, color='b', label=BlueName, ax=ax1)
    
    plt.title(Title)
    plt.xlabel("Price of Cars")
    plt.ylabel("Proportion of Cars")
    
    plt.show()
    plt.close()
    
    
```

```python
#Polly plot or Polynomial Regression Plots

def PollyPlot (xtrain, xtest, ytrain, ytest, lr, poly_transform):
    height=10
    width=12
    plt.figure(figsize=(width, height))
    
    xmax= max([xtrain.values.max(), xtest.values.max()])
    xmin=min([xtrain.values.min(), xtest.values.min()])
    x= np.arange(xmin, xmax, 0.1)
    
    plt.plot(xtrain, ytrain, "ro", label="Training Data")
    plt.plot(xtest, ytest, "go", label="Testing Data")
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1,1))))
    
    plt.ylabel("Price")
    plt.legend()
    
    
    
```

We will now create something cool that can help us play with different polynomial degrees and the ratio of training and testing data 

```python
def f(order, test_data):
    x_train, x_test, y_train, y_test = train_test_split(xdata, ydata, test_size=test_data, random_state=0)
    pr = PolynomialFeatures(degree=order)
    x_train_pr = pr.fit_transform(x_train[['horsepower']])
    x_test_pr = pr.fit_transform(x_test[['horsepower']])
    poly = LinearRegression()
    poly.fit(x_train_pr,y_train)
    PollyPlot(x_train[['horsepower']], x_test[['horsepower']], y_train,y_test, poly, pr)
```
